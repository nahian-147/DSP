```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from scipy.io.wavfile import read
```

```python
voice0 = read('hello_world0.wav')
voice1 = read('hello_world1.wav')
```

```python
plt.figure(figsize=(20,5),dpi=200)
plt.subplot(2,1,1)
plt.plot(voice0[1],'blue')
plt.subplot(2,1,2)
plt.plot(voice1[1],'green')
```

```python
N = len(voice0[1])
T = 1/voice0[0]
N*T
```

```python
FT = fft(voice0[1])
xfm = fftfreq(N, T)[:N//2]
plt.plot(xfm, 2.0/N * np.abs(FT[0:N//2]))
```

```python
plt.figure(figsize=(20,5),dpi=200)
plt.subplot(2,1,1)
N = len(voice0[1])
T = 1/voice0[0]
N*T
FT = fft(voice0[1])
xfm = fftfreq(N, T)[:N//2]
plt.plot(xfm, 2.0/N * np.abs(FT[0:N//2]),'blue')
plt.subplot(2,1,2)
N = len(voice0[1])
T = 1/voice1[0]
N*T
FT = fft(voice1[1])
xfm = fftfreq(N, T)[:N//2]
plt.plot(xfm, 2.0/N * np.abs(FT[0:N//2]),'blue')
```
