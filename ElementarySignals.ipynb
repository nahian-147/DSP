```python
import numpy as np
import matplotlib.pyplot as plt
```

### Impulse Function

```python
x = np.arange(-25,26)
```

```python
y = np.zeros(51)
y[25] = 1
```

```python
plt.figure(figsize=(10,2.5),dpi=200)
plt.bar(x,y,width=0.05)
plt.scatter(x,y,s=2,color='red')
```

### Unit Step Signal 

```python
x = np.arange(51)
y = [0 for i in range(25)] + [1 for i in range(26)]
y = np.array(y)
```

```python
plt.figure(figsize=(10,2.5),dpi=200)
plt.bar(x,y,width=0.15)
plt.scatter(x,y,s=1)
```

### Ramp Signal 

```python
x = np.arange(-25,26)
y = np.zeros(51)
for k in range(25,51):
    y[k] = k-25
```

```python
plt.figure(figsize=(10,2.5),dpi=200)
plt.bar(x,y,width=0.15)
plt.scatter(x,y,s=1)
```

### Exponential Signal

```python
x = np.arange(51)
y = np.exp(0.081*x)
```

```python
plt.figure(figsize=(10,2.5),dpi=200)
plt.bar(x,y,width=0.15)
plt.scatter(x,y,s=1)
```
